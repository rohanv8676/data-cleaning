# Cleaning the Data

#### Converting so it has 5 unique positions (PG, SG, SF, PF, C) 

```python
df['Pos'].replace('PG-SG', 'PG', inplace=True)
df['Pos'].replace('PG-SF', 'PG', inplace=True)
df['Pos'].replace('SG-SF', 'SG', inplace=True)
df['Pos'].replace('SG-PG', 'SG', inplace=True)
df['Pos'].replace('SG-PF', 'SG', inplace=True)
df['Pos'].replace('SF-SG', 'SF', inplace=True)
df['Pos'].replace('SF-PF', 'SF', inplace=True)
df['Pos'].replace('SF-C', 'SF', inplace=True)
df['Pos'].replace('PF-C', 'PF', inplace=True)
df['Pos'].replace('PF-SF', 'PF', inplace=True)
df['Pos'].replace('C-PF', 'C', inplace=True)
```

#### Creating a new dataframe for cleaned data

```python
cleaned_df = pd.DataFrame(columns = df.columns)
cleaned_df = cleaned_df.drop(['Rk', 'Age', 'Season', 'MVP'], axis=1)
```

#### Creating a list of all unique NBA players

```python
players = df['Player'].unique()
players
```

#### Inserting the mean statistics of each unique player

```python
for i in range(len(players)):
    current_player = players[i]
    player_stats = df[df['Player'] == current_player]
    
    cleaned_df.loc[i, 'Player'] = current_player
    cleaned_df.loc[i, 'Pos'] = player_stats['Pos'].value_counts().idxmax()
    cleaned_df.at[i, 'Tm'] = player_stats['Tm'].unique()
    cleaned_df.loc[i, 'G'] = round(player_stats['G'].mean(), 1)
    cleaned_df.loc[i, 'GS'] = round(player_stats['GS'].mean(), 1)
    cleaned_df.loc[i, 'MP'] = round(player_stats['MP'].mean(), 1)
    cleaned_df.loc[i, 'FG'] = round(player_stats['FG'].mean(), 1)
    cleaned_df.loc[i, 'FGA'] = round(player_stats['FGA'].mean(), 1)
    cleaned_df.loc[i, 'FG%'] = round(player_stats['FG%'].mean() * 100, 1)
    cleaned_df.loc[i, '3P'] = round(player_stats['3P'].mean(), 1)
    cleaned_df.loc[i, '3PA'] = round(player_stats['3PA'].mean(), 1)
    cleaned_df.loc[i, '3P%'] = round(player_stats['3P%'].mean() * 100, 1)
    cleaned_df.loc[i, '2P'] = round(player_stats['2P'].mean(), 1)
    cleaned_df.loc[i, '2PA'] = round(player_stats['2PA'].mean(), 1)
    cleaned_df.loc[i, '2P%'] = round(player_stats['2P%'].mean() * 100, 1)
    cleaned_df.loc[i, 'eFG%'] = round(player_stats['eFG%'].mean() * 100, 1)
    cleaned_df.loc[i, 'FT'] = round(player_stats['FT'].mean(), 1)
    cleaned_df.loc[i, 'FTA'] = round(player_stats['FTA'].mean(), 1)
    cleaned_df.loc[i, 'FT%'] = round(player_stats['FT%'].mean() * 100, 1)
    cleaned_df.loc[i, 'ORB'] = round(player_stats['ORB'].mean(), 1)
    cleaned_df.loc[i, 'DRB'] = round(player_stats['DRB'].mean(), 1)
    cleaned_df.loc[i, 'TRB'] = round(player_stats['TRB'].mean(), 1) 
    cleaned_df.loc[i, 'AST'] = round(player_stats['AST'].mean(), 1)
    cleaned_df.loc[i, 'STL'] = round(player_stats['STL'].mean(), 1)
    cleaned_df.loc[i, 'BLK'] = round(player_stats['BLK'].mean(), 1)
    cleaned_df.loc[i, 'TOV'] = round(player_stats['TOV'].mean(), 1)
    cleaned_df.loc[i, 'PF'] = round(player_stats['PF'].mean(), 1)
    cleaned_df.loc[i, 'PTS'] = round(player_stats['PTS'].mean(), 1)
    cleaned_df.loc[i, 'Age at First Season'] = player_stats['Age'].min()
    cleaned_df.loc[i, 'Age at Last Season'] = player_stats['Age'].max()
    cleaned_df.loc[i, 'Num Season(s)'] = len(player_stats['Season'].unique())
    cleaned_df.loc[i, 'MVP(s)'] = player_stats['MVP'].values.sum()

cleaned_df = cleaned_df.astype({'Age at First Season' : int, 'Age at Last Season' : int, 'Num Season(s)' : int, 'MVP(s)' : int})
```

```python
cleaned_df.to_csv('cleaned_NBA_Player_Stats.csv')
```
